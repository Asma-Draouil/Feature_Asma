```python
# Fonction pour enlever les caractères spéciaux
def enlever_caracteres_speciaux(texte):
    return re.sub(r'[^A-Za-z0-9À-ÿ\s]', '', texte)
```

```python
# Appliquer la fonction à toutes les colonnes du DataFrame
df = df.applymap(lambda x: enlever_caracteres_speciaux(str(x)) if isinstance(x, str) else x)
```
