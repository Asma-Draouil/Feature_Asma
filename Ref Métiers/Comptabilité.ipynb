```python
# URL du site web
url = 'http://rtmc.emploi.nat.tn/dm/index.php/fichemetier/rtmc_resp/M1203'
# Charger le contenu HTML depuis la page web
response = requests.get(url)
response.raise_for_status()  # Vérifier que la requête s'est bien passée
html = response.text
```

```python
# Find the table in the HTML
url ='http://rtmc.emploi.nat.tn/dm/index.php/fichemetier/rtmc_resp/M1203'
# Charger le contenu HTML depuis la page web
response = requests.get(url)
response.raise_for_status()  # Vérifier que la requête s'est bien passée
html = response.text

table = soup.findall("table", class_="table table-sm",attrs={'href':'#competences'})

# Find all rows in the table
rows = table.find_all("tr")

# Extract the text from each row (ignoring the <i> tags)
data = []
for row in rows:
    columns = row.find_all("td")
    if len(columns) > 1:
        # Extracting the text of the second column (index 1)
        data.append(columns[1].get_text(strip=True))

print(data)
```

```python
url ='http://rtmc.emploi.nat.tn/dm/index.php/fichemetier/rtmc_resp/M1203/competences'
# Charger le contenu HTML depuis la page web
response = requests.get(url)
response.raise_for_status()  # Vérifier que la requête s'est bien passée
html = response.text
# Target the specific parent div
parent_div = soup.find("div", class_="accordion-body__contents")

# Find the table within this div
table = parent_div.find("table", class_="table table-sm")

# Extract all rows within the table
rows = table.find_all("tr")

# Extract and store the text from each row
row_texts = [row.find("td", style="padding-left:0").get_text(strip=True) for row in rows]

# Print the extracted rows as a list
print(row_texts)
```
